## IMPORT DEI DATASET

```python
inventory_items_raw = pd.read_csv('inventory.csv') #insert the path
```

## Dataset Inventory_items

```python
inventory_items_raw.info()
```

```python
inventory_items = inventory_items_raw.copy()
```

```python
inventory_items = inventory_items.rename(columns={
                                      "id": "inventory_item_id",
                                      "created_at" : "item_created_at",
                                      "sold_at" : "item_sold_at",
                                      "cost" : "product_cost",
                                      "name" : "product_distribution_center"})
```

```python
inventory_items = inventory_items.drop(['product_sku', 'product_distribution_center_id', 'id_1', 'latitude', 'longitude'], axis=1)
```

```python
inventory_items.info()
```

Notiamo che abbiamo dei missing value nelle seguenti colonne:
- product_name
- product_brand

```python
inventory_items.head(1)
```

```python
inventory_items['sold'] = inventory_items['item_sold_at'].copy()
inventory_items['sold'] = inventory_items['sold'].fillna(0)
inventory_items['sold'][inventory_items['sold']!=0] = 1
```

```python
inventory_items.head()
```

```python
inventory_items["year_item_sold_at"] = [str(i)[0:4] for i in inventory_items["item_sold_at"]]
inventory_items["mounth_item_sold_at"] = [str(i)[5:7] for i in inventory_items["item_sold_at"]]
inventory_items["year_item_created_at"] = [str(i)[0:4] for i in inventory_items["item_created_at"]]
inventory_items["mounth_item_created_at"] = [str(i)[5:7] for i in inventory_items["item_created_at"]]
```

```python
inventory_items['year_item_sold_at'] = inventory_items['year_item_sold_at'].str.replace('nan','0')
inventory_items['year_item_created_at'] = inventory_items['year_item_created_at'].str.replace('nan','0')
inventory_items["year_item_sold_at"] = inventory_items["year_item_sold_at"].astype(int)
inventory_items["year_item_created_at"] = inventory_items["year_item_created_at"].astype(int)
```
