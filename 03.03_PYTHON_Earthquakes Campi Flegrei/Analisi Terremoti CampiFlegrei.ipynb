## INTRODUCTION

### ABOUT - Cosa sono i Campi Flegrei?

I Campi Flegrei sono un'**area vulcanica attiva** situata ad ovest di Napoli, che include i comuni di Bacoli, Monte di Procida, Pozzuoli, Quarto, Giugliano in Campania e parte della città di Napoli. Quest'area è caratterizzata da un'alta **densità abitativa**, dove vivono più di 500.000 persone. Il nome Campi Flegrei, dal greco letteramente *"campi ardenti"*, denota la natura vulcanica dell'area e la presenza di numerose fumarole e acque termali, ben note e sfruttate nell’antichità.

A differenza del più noto Vesuvio, i Campi Flegrei non sono caratterizzati da un unico edificio vulcanico principale, ma sono piuttosto un campo vulcanico attivo da più di 80.000 anni, con diversi centri vulcanici situati all'interno e in prossimità di un'area depressa chiamata caldera. La **caldera** è il risultato del ripetuto sprofondamento di una vasta area provocato dal collasso del tetto del serbatoio magmatico superficiale a seguito dello svuotamento dello stesso per opera di almeno due grandi eruzioni: l'Ignimbrite Campana (40.000 anni) e il Tufo Giallo Napoletano (15.000 anni). L'eruzione dell'Ignimbrite Campana è l'eruzione a più elevata energia conosciuta nel Mediterraneo: con essa un'enorme quantità di cenere è stata dispersa nell'atmosfera, influenzando il clima non solo a livello regionale ma probabilemte anche a livello mondiale.
 
Dopo l'eruzione del Tufo Giallo Napoletano l'attività vulcanica dei campi Flegrei è stata particolarmente intensa con più di 27 eruzioni solo negli ultimi 5.500 anni, l'ultima delle quali, avvenuta nel **1538**, ha generato il cono di tufo di **Monte Nuovo**.

I Campi flegrei hanno un **diamentro** di circa 15 km, mentre quello del Vesuvio è di circa 4km. 

<center><img src="img/Campi-Flegrei-e-Vesuvio.jpg" width="500" height="400" /></center>

L'attività eruttiva degli ultimi 15.000 anni, successiva quindi all'eruzione del Tufo Giallo Napoletano,  è stata principalmente esplosiva. Tale attività è stata interrotta da due lunghi periodi di quiescenza che hanno consentito di suddividirla in tre epoche differenti:

- prima Epoca di attività (15.000-10.600 anni)
- seconda Epoca di attività (9.600 - 9.200 anni)
- terza Epoca di attività (5.500 - 3.800 anni)

In totale, nel corso delle tre Epoche sono state individuate circa **70 eruzioni**.

La caldera dei Campi Flegrei è soggetta a lenta deformazione del suolo nota con il nome locale di **bradisismo**. Nei periodi **1970-72** e **1982-84** l’area flegrea è stata interessata da **crisi bradisismiche** in cui il suolo, nell'abitato di Pozzuoli in particolare, ha subito un sollevamento totale massimo di circa 3.5 m. La prima crisi causò l'abbandono forzato dell'area fatiscente di Rione Terra; la seconda crisi in particolare fu caratterizzata da intensa sismicità con gravi danni agli edifici. Dopo le crisi si è avuto un periodo di generale subsidenza, interrotta a partire dal **2005** da un'inversione del fenomeno che ha portato ad un costante sollevamento del suolo, al momento ancora in atto.

### PURPOSE

Dal momento che il territorio dei Campi Flegrei si sta avviando a vivere una nuova crisi bradisimica, dopo le più recenti degi anni '70 e '80, nei giorni d'oggi vi è la necessità di comprendere cosa sta accadendo in quest'area e farlo in maniera tempestiva.

L'obiettivo di questa ricerca, oltre ad approfondire la conoscenza del fenomeno, è quella di avere a disposizione un quadro approssimativo dell'evoluzione in tempo reale, in attesa della pubblicazione ufficiale dei bollettini degli esperti dell'INGV (Istituto Nazionale Geofisica e Vulcanologia).

Pertanto, prima di tutto, con questo notebook analizzo i dati storici per evidenziare quali sono le tendenze del fenomeno. Dopo di che, realizzo una dashboard con i principali KPI al fine di poter eseguire un monitoraggio in tempo reale del fenomeno con i dati a disposizione.

###  IMPORT LIBRARY

```python
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from mpl_toolkits.mplot3d import axes3d
import plotly.express as px
```

### IMPORT DATA

```python
# import dati anni dal 2000 al 2021

df_2000_2021 = pd.read_csv('dataset\dati_fle_2000_2021.csv')

df_2000_2021 = df_2000_2021.reset_index()

df_2000_2021 = df_2000_2021.rename(columns={"index": "Time",
                                            "Tempo origine (UTC)": "Magnitude", 
                                            "Magnitudo": "Latitude", 
                                            "Latitudine N": "Longitude", 
                                            "Longitudine E": "Depth/km",  
                                            "Profondità (km)": "Note"
                                            })

df_2000_2021 = df_2000_2021.drop('Note', axis=1)

df_2000_2021 = df_2000_2021[['Time','Latitude','Longitude','Depth/km','Magnitude']]
```

```python
# import dati anni 2022-2023

df_2022 = pd.read_csv('dataset\events 2022.csv')
df_2023 = pd.read_csv('dataset\events 2023.csv')
```

### JOIN DATA

```python
# creo un unico df che comprendere anni 2000-2023

df = pd.concat([df_2022,df_2023])
df = df.reset_index()
df = df.drop('index', axis=1)

df = df.drop(['#EventID','MagType','MagErr', 'Area', 'Type', 'Level'], axis=1)

df = pd.concat([df,df_2000_2021])
df = df.sort_values(by='Time')
df = df.reset_index()
df = df.drop('index', axis=1)
```

### DATA CLEANING

```python
# convertion data type
df['Time'] = df['Time'].str.replace('-','/')
df['Time'] = pd.to_datetime(df['Time'],format='mixed')
```

```python
# create new columns
df['year'] = df['Time'].dt.year
df['month'] = df['Time'].dt.month

# convertion data type
df['year'] = df['year'].astype(str)
df['month'] = df['month'].astype(str)
```

```python
# create new columns
df['year_month']= pd.to_datetime(df.Time).dt.strftime('%Y/%m')
df['date']= pd.to_datetime(df.Time).dt.strftime('%Y/%m/%d')
```

```python
# elimino righe di novembre 2023, mese per cui al momento in cui si svolge l'analsi non si hanno a disposizione tutti i dati
df = df[df['year_month']!='2023/11']
```

### Earthquakes Locations

I terremoti sono concentrati nell'area delimitata da Lucrino, Cigliano (Pozzuoli), Bagnoli (Napoli) e dal mare (Golfo di Pozzuoli), il cui centro è il porto di Pozzuoli.

La maggior parte dei terremoti si verifica nell'area sinteticamente identificata come "Solfatara-Piscarelli", ma più precisamente si tratta dell'area delimitata Solfatara, Pisciarelli, Antiniana, Bagnoli e dal porto di Pozzuoli.

Si nota che nella zona tra Pozzuoli e Bagnoli, nei pressi di via Napoli, vi è una particolare concentrazione di terremoti molto superficiali (entro circa 500m di profondità).

```python
# realizzo una mappa satellitare con la localizzazione di ogni terremoto
# dove la grandezza di ogni punto indica la magnitudo, mentre la scala di colore indica la profondità

color_scale = [(0, 'yellow'), (1,'red')]

fig = px.scatter_mapbox(df[df['Magnitude']>=0],
                        hover_name='Time', 
                        lat='Latitude', 
                        lon='Longitude',
                        color='Depth/km',
                        color_continuous_scale=color_scale,
                        size='Magnitude',
                        zoom=11, 
                        height=500,
                        width=800
                        )

fig.update_layout(mapbox_style="carto-positron")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
plt.scatter(x=df['Longitude'], y=df['Latitude'])
plt.title("Eartquakes locations distribution")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(linestyle='--')
plt.show()
```

Tramite un grafico tridimensionale si rappresenta la posizione nell'area e nel sottosuolo dei terremoti. La grandezza ed il colore di ogni sfera rappresentano la magnitudo del terremoto.

```python
x = df.Longitude
y = df.Latitude
z = df['Depth/km']
c = df.Magnitude
figure(figsize=(12,12))
ax = plt.axes(projection='3d')
ax.invert_zaxis()
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.set_zlabel("Depth")
ax.scatter(x, y, z, s=(c*200), c=-c, linewidth=1.5);
#ax.view_init(180, 45) # per ruotare grafico
plt.show()
```

    - La frequenza di terremoti con magnitudo superiore al 2 è aumentata nel 2023 rispetto a quanto avvenuto nei precedenti 23 anni, infatti nel periodo tra agosto del 2000 e dicembre del 2022 del totale dei terremoti verificatesi (6.560) lo 0.671% ha avuto una magnitudo superiore al 2, mentre in quello tra gennaio del 2023 e ottobre del 2023 tale percentuale è 1.183% con 5.831 terromoti totali registrati;
    - La magnitude massima mensile sta aumentando, infatti negli ultimi ultimi 12 mesi si è verificato sempre almeno un terremoto di magnitudo superiore a 2.5 in ogni mese, ad eccezione del mese di luglio 2023 quando il massimo raggiunto è stato 1.86;
    - Negli ultimi 12 mesi la media mensile di terremoti con magnitudo superiore al 2.5 è stata di circa 3 eventi al mese;
    - La magnitudine massima registata negli ultimi 40 anni è stata pari a 4.2 in data 27 settembre 2023;
- DEPTH
    - La maggior parte dei terremoti si è verificata nei primi 3km del sottosuolo (il 75% entro i 2.15km), con particolare concentrazione nell'intervallo tra 1km e 3km;
    - Negli ultimi anni, mentre la profondità minima annuale si sta mantenendo al di sotto di 1km, quella massima invece sta aumentando ed è arrivata a superare i 4km;
    - Negli ultimi 12 mesi la profondità massima media è stata di 4.17km mentre quella minima di 0.41km;
- CORRELATIONS
    - Il 40% dei terremoti registrati si sono verificati nei primi 3km di profondità con una magnitudo inferiore al 2; 
    - Vi è una correlazione positiva tra il numero di terremoti giornalieri e la magnitudo massima raggiunta nella giornata (ossia nelle giornate con un maggior numero di terremento tende a verificarsi almeno un terremoto con alta magnitudo).


In sintesi, nell'ultimi mesi si sta verificando un inteso aumento del fenomeno bradisismico nei Campi Flegrei. Questo è evidenziato da un incremento della frequenza di terrementi e da un aumento delle magnitudo superiori al 2 registrate. Inoltre, è emerso che vi sono delle area con particolare concentrazione di terremoti, ossia quella deliminata da Solfatara, Pisciarelli, Antiniana, Bagnoli e dal porto di Pozzuoli.
